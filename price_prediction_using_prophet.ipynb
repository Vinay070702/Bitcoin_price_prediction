```python
#read the data
df = pd.read_csv('BTC-USD_train.csv')
#show the data
df
```

```python
#select the date and price 
df = df[['Date','Close']]
#rename thee features:
df = df.rename(columns={'Date':'ds','Close':'y'})
#show the data
df
```

```python
#getting the last 20 rows of data and store into a new variable 
last = df[len(df)-20:]
last
```

```python
# Give all of the rows i the data set expect for the last 20 rows of the data
df = df[:-20]
df
```

```python
#Getting the facebook prophet library
from prophet import Prophet
#Create the Prophet object(model)
fpb = Prophet(daily_seasonality = True)
#Fit/train the model
fpb.fit(df)
future = fpb.make_future_dataframe(periods = 365)
forecast = fpb.predict(future)
```

```python
#import the facebooks prophet plotting libraries
from prophet.plot import plot_plotly
#plot the data 
plot_plotly(fpb,forecast)
```

```python
#show the model predictions 
forecast[forecast.ds == '2021-08-12']['yhat']
```

```python
#show the actual value of the coin at that day
last[last.ds=='2021-08-12']['y']
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(df['y'],forecast['yhat'][:-365])
print('r2 score for perfect model is', r2)
```
